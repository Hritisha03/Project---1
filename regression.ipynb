```python
#LINEAR REGRESSION
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([[1000], [1500],[2200],[3400]]) # X needs to be a 2D array for scikit-learn
Y = np.array([200,150,390,430]) #"Train" the model (This finds the best weights 'w' and bias 'b')

model = LinearRegression()#Initialize the Linear Regression Model
model.fit(X,Y)

#See the learned parameters
print(f"Leanrned Weight (w): {model.coef_[0]:.4f}")
print(f"Learned Bias (b): {model.intercept_:.4f}")

#Make a prediction for a new house
new_house = np.array([[1560]])
predicted_price = model.predict(new_house)
print(f"Predicted price for the required house: ${predicted_price[0]:.2f}k")

plt.scatter(X, Y, color='blue', label='Actual Data')
X_line = np.array([[1000], [3400]])
Y_line = model.predict(X_line)
plt.plot(X_line, Y_line, color='red', linewidth=2, label='Linear Regression Line')

plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($k)')
plt.title('Your Custom Linear Regression Model')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures

# 1. Your custom data points
X = np.array([[1000], [1500], [2200], [3400]])
Y = np.array([200, 150, 390, 430])

# Transform to a 3rd-degree polynomial (Guaranteed to overfit without help)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

# 2. Train Ridge (L2) and Lasso (L1) with an Alpha penalty of 1.0
ridge_model = Ridge(alpha=1.0)
lasso_model = Lasso(alpha=1.0, max_iter=10000) # Added max_iter for convergence

ridge_model.fit(X_poly, Y)
lasso_model.fit(X_poly, Y)

# 3. Check out the learned weights (coefficients)
print("Ridge Weights:", ridge_model.coef_)
print("Lasso Weights:", lasso_model.coef_)
```
